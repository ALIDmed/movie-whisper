```python
# director_name, actor_2_name, genres, actor_1_name, movie_title, actor_3_name, title_year
df = df[['movie_title', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres']]
```

```python
df.head()
```

```python
df.genres = df.genres.str.replace("|", " ")
```

```python
# remove last character from movie title
df.movie_title = df.movie_title.apply(lambda x: x[:-1])
```
